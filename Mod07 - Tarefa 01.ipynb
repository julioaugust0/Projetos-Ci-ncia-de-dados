#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável. 
    Dica: 
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis, 
        3. conte o número de categorias dessa variável 
        4. acumule essa informação de 3. na lista que você criou em 1. 
        5. No final, essa lista pode ser atribuída à nossa variável.
3. Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.

````python
import pandas as pd
import os

# Defina o caminho para o arquivo (caminho absoluto)
file_path = 'C:/Users/julio/Downloads/demo01.csv'  # Use a abordagem que preferir

# Verifique se o arquivo existe no caminho especificado
if os.path.isfile(file_path):
    # Carregar a base de dados
    df = pd.read_csv(file_path)

    # Passo 1: Criar DataFrame de Metadados
    metadados = pd.DataFrame({
        'nome_variavel': df.columns,
        'tipo_dado': df.dtypes
    })

    # Passo 2: Adicionar Coluna "qtd_categorias"
    qtd_categorias = []
    for col in df.columns:
        if df[col].dtype == 'object' or df[col].nunique() < 10:
            qtd_categorias.append(df[col].nunique())
        else:
            qtd_categorias.append(None)

    metadados['qtd_categorias'] = qtd_categorias

    print("Metadados:")
    print(metadados)

    # Passo 3: Criar Variáveis Dummy
    variaveis_qualitativas = metadados[(metadados['tipo_dado'] == 'object') | (metadados['qtd_categorias'].notnull())]['nome_variavel']

    df_dummies = pd.get_dummies(df, columns=variaveis_qualitativas, drop_first=True)

    print("\nDataFrame com variáveis dummy:")
    print(df_dummies.head())
else:
    print(f"Arquivo {file_path} não encontrado. Verifique o caminho e tente novamente.")
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
import pandas as pd
import os

# Defina o caminho para o arquivo (caminho absoluto)
file_path = 'C:/Users/julio/Downloads/demo01.csv'  # Use a abordagem que preferir

# Verifique se o arquivo existe no caminho especificado
if os.path.isfile(file_path):
    # Carregar a base de dados
    df = pd.read_csv(file_path)

    # Identificar variáveis qualitativas que precisam de dummies
    metadados = pd.DataFrame({
        'nome_variavel': df.columns,
        'tipo_dado': df.dtypes
    })
    
    qtd_categorias = []
    for col in df.columns:
        if df[col].dtype == 'object' or df[col].nunique() < 10:
            qtd_categorias.append(df[col].nunique())
        else:
            qtd_categorias.append(None)

    metadados['qtd_categorias'] = qtd_categorias

    variaveis_qualitativas = metadados[(metadados['tipo_dado'] == 'object') | (metadados['qtd_categorias'].notnull())]['nome_variavel']

    # Criar variáveis dummy
    df_dummies = pd.get_dummies(df, columns=variaveis_qualitativas, drop_first=True)

    # Criar um DataFrame apenas com variáveis apropriadas para entrada no scikit-learn
    df_final = df_dummies.select_dtypes(include=[int, float, bool])

    print("\nDataFrame com variáveis apropriadas para entrada no scikit-learn:")
    print(df_final.head())
else:
    print(f"Arquivo {file_path} não encontrado. Verifique o caminho e tente novamente.")
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
import pandas as pd
import os

# Defina o caminho para o arquivo (caminho absoluto)
file_path = 'C:/Users/julio/Downloads/demo01.csv'  # Use a abordagem que preferir

# Verifique se o arquivo existe no caminho especificado
if os.path.isfile(file_path):
    # Carregar a base de dados
    df = pd.read_csv(file_path)
    
    # Tabela cruzada para possui_email em relação a mau
    tabela_email = pd.crosstab(df['possui_email'], df['mau'], normalize='index')
    print("Tabela cruzada para possui_email:")
    print(tabela_email)
    
    # Proporção de mau = 1 para cada categoria de possui_email
    proporcao_email = tabela_email.loc[:, 1] if 1 in tabela_email.columns else pd.Series([0] * len(tabela_email.index), index=tabela_email.index)
    print("\nProporção de mau = 1 para possui_email:")
    print(proporcao_email)
    
    # Tabela cruzada para posse_de_veiculo em relação a mau
    tabela_veiculo = pd.crosstab(df['posse_de_veiculo'], df['mau'], normalize='index')
    print("\nTabela cruzada para posse_de_veiculo:")
    print(tabela_veiculo)
    
    # Proporção de mau = 1 para cada categoria de posse_de_veiculo
    proporcao_veiculo = tabela_veiculo.loc[:, 1] if 1 in tabela_veiculo.columns else pd.Series([0] * len(tabela_veiculo.index), index=tabela_veiculo.index)
    print("\nProporção de mau = 1 para posse_de_veiculo:")
    print(proporcao_veiculo)
    
    # Comparação das variáveis
    print("\nComparação das proporções:")
    print(f"Proporção de mau = 1 para possui_email:\n{proporcao_email}")
    print(f"\nProporção de mau = 1 para posse_de_veiculo:\n{proporcao_veiculo}")
    
    # Determinar a variável mais poderosa
    if proporcao_email.max() > proporcao_veiculo.max():
        print("\nA variável 'possui_email' parece ser mais poderosa para prever a probabilidade de mau = 1.")
    else:
        print("\nA variável 'posse_de_veiculo' parece ser mais poderosa para prever a probabilidade de mau = 1.")
else:
    print(f"Arquivo {file_path} não encontrado. Verifique o caminho e tente novamente.")
````
